# Define method for generating text from files

```python
def text_generator(files, splitter=None):
    # splitter is responsible for breaking up text into smaller sequences
    # i.e. perhaps by splitting on \n
    for f in files:
        text = open(f).read()
        # remove highlights
        text = text.split('@highlight')[0]
        if splitter is None:
            text = [text]
        else:
            text = splitter(text)
        for t in text:
            if not t or len(t) < 10:
                continue
            yield t
```

# Define batch generator

```python
splitter = lambda x: x.split('\n')
```

```python
def sequencer(tokens):
    return [tokens[:i] for i in range(1, len(tokens)+1)]
```

```python
list(sequencer('a quick brown fox'))
```

```python
import random
import numpy as np

def batch_generator(files, maxlen, splitter=None, sequencer=None):
    while True:
        random.shuffle(files)
        for text in text_generator(files, splitter):
            tokens = tokenizer.texts_to_sequences([text])[0]
            sequences = sequencer(tokens)
            if len(sequences) < 3:
                continue
            
            X = pad_sequences(sequences, maxlen=maxlen)

            y = tokenizer.sequences_to_matrix([[i] for x in X for i in x])
            y = y.reshape((len(sequences), maxlen, tokenizer.num_words))
            
            # offset
            X = [X[:-1], X[:-1]]
            y = y[1:]
            
            # include text for debugging only
            if len(X[0]) <= 100:
                yield text, X, y
            else:
                # break up batches from large sentences to avoid resource exhaustion
                yield text, [X[0][:100], X[1][:100]], y[:100]
                yield text, [X[0][100:], X[1][100:]], y[100:]
```

# Training

```python
n_heads = 8
n_layers = 6
d_model = 64*n_heads
vocab_size = tokenizer.num_words
sequence_len = 30
```

```python
steps_per_epoch = 0
for text in text_generator(files, splitter):
    steps_per_epoch += 1
```

```python
steps_per_epoch
```

```python
from keras.callbacks import TerminateOnNaN
callbacks = [TerminateOnNaN()]
```

```python
batch_gen = batch_generator(files, sequence_len, splitter=splitter, sequencer=sequencer)
```

```python
from collections import deque
# trick that allows us to look at the most recently trained on batch
# helpful post-mortem if training terminates on NaN (batches[0])
batches = deque(maxlen=1)
train_gen = ((batches.append(x), x[1:])[-1] for x in batch_gen)
```

```python
from model import Transformer
model = Transformer(
        n_heads=n_heads, encoder_layers=n_layers, decoder_layers=n_layers,
        d_model=d_model, vocab_size=vocab_size, sequence_len=sequence_len,
        layer_normalization=True, dropout=True,
        residual_connections=True)
```

```python
model.summary()
```

```python
loss = 'categorical_crossentropy'
```

```python
from keras.optimizers import adam
model.compile(loss=loss, optimizer=adam(lr=1e-5))
```

```python
n_epochs = 100
model.fit_generator(train_gen, steps_per_epoch=steps_per_epoch, epochs=n_epochs,
                    callbacks=callbacks)
```
